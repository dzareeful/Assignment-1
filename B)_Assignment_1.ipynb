```python
import cv2
# from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Open image using openCV2
normal = cv2.imread("normal.jpg")
grind = cv2.imread("grind 2.jpg")
puzzled = cv2.imread("puzzled.jpg")

normal = cv2.cvtColor(normal, cv2.COLOR_BGR2RGB)
grind = cv2.cvtColor(grind, cv2.COLOR_BGR2RGB)
puzzled = cv2.cvtColor(puzzled, cv2.COLOR_BGR2RGB)

normalcp = np.copy(normal)
grindcp = np.copy(grind)
puzzledcp = np.copy(puzzled)


plt.figure(figsize=[25,5])
plt.subplot(141);plt.imshow(normal);plt.title("Normal");
plt.subplot(142);plt.imshow(grind);plt.title("Grind");
plt.subplot(143);plt.imshow(puzzled);plt.title("Puzzled");
```

```python
# initialize the Haar Cascade face detection model
face_cascade = cv2.CascadeClassifier(cv2.samples.findFile(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'))

faces = face_cascade.detectMultiScale(normal)

#face = x,y,z
for (x,y,w,h) in faces:
  #imgView = cv2.rectangle(imgView,(x,y),(x+w,y+h),(255,0,0),1)
  normalcp[y:y+h,x:x+w] = cv2.medianBlur(normalcp[y:y+h,x:x+w],35)
  grindcp[y:y+h,x:x+w] = cv2.medianBlur(grindcp[y:y+h,x:x+w],35)
  puzzledcp[y:y+h,x:x+w] = cv2.medianBlur(puzzledcp[y:y+h,x:x+w],35)


plt.figure(figsize=[7,10])
plt.subplot(321);plt.imshow(normal);plt.title("Normal");
plt.subplot(322);plt.imshow(normalcp);plt.title("Blurred Normal");
plt.subplot(323);plt.imshow(grind);plt.title("Grind");
plt.subplot(324);plt.imshow(grindcp);plt.title("Blurred Grind");
plt.subplot(325);plt.imshow(puzzled);plt.title("Puzzled");
plt.subplot(326);plt.imshow(puzzledcp);plt.title("Blurred Puzzled");

#plt.imshow(normal, cmap="gray")
#plt.show()
```
